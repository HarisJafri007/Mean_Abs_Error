Import Pandas Library to handle Data.

```python
import pandas as pd
```

Save the file path in a variable.

```python
file_path = 'melb_data.csv'
```

Use pd.read_csv to extract the data and store it inside a Variable as a DataFrame.

```python
house_data = pd.read_csv(file_path)
```

Check the first five rows of the DataFrame.

```python
house_data.head()
```

Notice that as per our directions, Python has only shown first five rows of the Data Set. Also notice that there are a total of 21 Columns.

The .describe() method in Pandas selects the Columns with Numerical Datatypes and shows the mean, spread, min & max values and count.

```python
house_data.describe()
```

Notice that count in Rooms and count in Year Built Column (and also some other Column) is different. This is because some values are missing i.e. Data is not entered. This causes the entire entry (row) to be effected. We may find the total rows before deleting the rows having missing entries!

```python
house_data.shape
```

There are a total of 13580 Rows and the usual 21 Columns in the Data set. We may remove the remove the rows with missing values and find the total rows again!

```python
house_data = house_data.dropna(axis=0)
```

```python
house_data.shape
```

Notice that number of rows in the data set has been reduced from 13580 to 6196. More than half of the rows have been deleted!

```python
house_data.head()
```

```python
house_data.describe()
```

Now we would like to predict the Price. The Variable y is used to store the DataFrame Column. 

```python
# prediction_target
y = house_data.Price
```

Now in order to store features in a Variable. We first must be able to view what are the available options.

```python
house_data.columns
```

House Prices are effected by a lot of factors but generally for simplicitity we may select the features to be Rooms, Bathrooms, Landsize, Lattitude and Longtitude.

```python
# features
house_features = ['Rooms','Bathroom','Landsize','Lattitude','Longtitude']
```

Storing the DataFrame with required features inside Variable X.

```python
X = house_data[house_features]
```

```python
X.head()
```

```python
X.describe()
```

Now we can import DecisionTreeRegressor and hold it to a random state and use it to train the Model using X and y.

```python
from sklearn.tree import DecisionTreeRegressor

#define model
house_model = DecisionTreeRegressor(random_state=1)

#fit model
house_model.fit(X,y)
```

```python
print('Making Prediction for the following 5 Houses are')
print(X.head())
print('The Predictions are', house_model.predict(X.head()))
```

Did you notice something? We train the model and then asked the model to predict the prices using features from the same data set we trained the Model on. How will we check how much efficient our Model is! Is it making any mistake? Well there are many methods to check the efficiency of our Model and one of them is mean_absolute_error

```python
predicted_home_prices = house_model.predict(X)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(y, predicted_home_prices)
mae
```

On Average, our predictions are off by about $1115.77

Another way of handling this issue is to split the data into test data and train data. Use the train data part to train the model and check the efficiency of the model by test data!

```python
from sklearn.model_selection import train_test_split
```

```python
train_X, val_X, train_y, val_y = train_test_split(X,y, random_state=0)
```

```python
house_model2 = DecisionTreeRegressor(random_state=1)
```

```python
house_model2.fit(train_X, train_y)
```

```python
predicted_X = house_model2.predict(val_X)
```

```python
mae2 = mean_absolute_error(val_y, predicted_X)
mae2
```

```python
mae2 - mae
```

Wow, when we train and test Model without splitting the Data the mean absolute error was only \\$1000 and when we trained and tested the model using different sample from the same data, the mean absolute error reached up to \\$272,000
